```python
from pybtex.database import parse_file

bib_data = parse_file('assets/literature.bib')
```

```python
bib_data.entries
```

```python
entry = list(bib_data.entries.items())[0]
person = list(entry[1].persons.items())[0][1]
str(person[0])
```

```python
entry_template = """
1. **[{title}]({url})**

    *{authors}*

    <u>{journal}</u> {year}
"""

with open('assets/literature.md', 'w') as f:
    for entry_key, entry in bib_data.entries.items():
        title = entry.fields['title']
        authors = [" ".join(person.first_names + person.middle_names + person.last_names) for person in entry.persons["author"]]
        journal = entry.fields.get('journal', "")
        year = entry.fields.get('year')
        url = entry.fields.get('url')
        publisher = entry.fields.get('publisher', "")
        booktitle = entry.fields.get('booktitle', "")

        journal = f"{booktitle}. {journal}. {publisher}".replace(". .", ".").replace("..", ".").strip(".").strip().strip(".")

        if journal:
            journal = f"{journal}."

        f.write(entry_template.format(
            title=title,
            authors=", ".join(authors),
            journal=journal,
            year=year,
            url=url
        ))
```
